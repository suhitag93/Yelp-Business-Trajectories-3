```python
import csv
import pandas as pd
business_file ='yelp_academic_dataset_business.csv'
tips_file ='yelp_academic_dataset_tip.csv'
reviews_file = 'yelp_academic_dataset_review.csv'
tips_fields = ['business_id','user_id','date']
business_fields = ['attributes_Price Range','name','business_id', 
                   'categories','review_count', 'stars', 'state']

business_dataframe = pd.read_csv(business_file, low_memory=False, usecols= business_fields)

tips_dataframe = pd.read_csv(tips_file, low_memory= False, usecols = tips_fields)

business_dataframe = pd.read_csv(business_file, low_memory=False, usecols= business_fields)

tips_dataframe = pd.read_csv(tips_file, low_memory= False, usecols = tips_fields)
```

```python
tip_dates  = tips_dataframe['date'].tolist() 
```

```python
year = []
month = []
for i in tips_dataframe.index:
    y, m, d =tip_dates[i].split('-')
    year.append(y)
    month.append(m)
```

```python
tips_dataframe["year"] = year
```

```python
tips_dataframe['month'] = month
```

```python
del tips_dataframe['month_tips']
```

```python
review_fields = ['business_id', 'user_id','date']
reviews_dataframe = pd.read_csv(reviews_file, skipinitialspace = True, usecols = review_fields)
```

```python
reviews_dataframe.head()
```

```python
reviews_dataframe = reviews_dataframe[reviews_dataframe.date.str.contains('20')]
```

```python
reviews_dataframe.index
```

```python
review_dates = reviews_dataframe['date'].tolist()
r_year = []
r_month = []
for i in reviews_dataframe['date'].index:
    y, m,d = reviews_dataframe['date'][i].split('-')
    r_year.append(y)
    r_month.append(m)
```

```python
reviews_dataframe['year'] = r_year
```

```python
reviews_dataframe['month'] = r_month
del reviews_dataframe['year_reviews']
del reviews_dataframe['month_reviews']
```

```python
reviews_dataframe.head()
```

```python
reviews_dataframe = reviews_dataframe.rename(columns = {'user_id':'user_reviews'})
```

```python
reviews_dataframe.head()
```

```python
reviews_grouped = pd.DataFrame(reviews_dataframe.
                               groupby(['business_id','year','month'])
                               ['user_reviews'].apply(list))
reviews_grouped.reset_index()
```

```python
tips_grouped = pd.DataFrame(tips_dataframe.
                               groupby(['business_id','year','month'])
                               ['user_tips'].apply(list))
```

```python
tips_count = []
for i in tips_grouped.index:
    tips_count.append(len(tips_grouped['user_tips'][i]))
```

```python
tips_grouped['tips_count']=tips_count
```

```python
review_count = []
for i in reviews_grouped.index:
    review_count.append(len(reviews_grouped['user_reviews'][i]))

```

```python
reviews_grouped['review_counts'] = review_count
reviews_grouped.head()
```

```python
#merging the tips and reviews

merged_tips_reviews = pd.merge(tips_grouped, reviews_grouped, how = 'outer',
                               on = ['business_id','year','month'],
                               right_index = True, left_index = False)
```

```python
merged_tips_reviews['review_counts'].fillna(0, inplace=True)
merged_tips_reviews['tips_count'].fillna(0, inplace=True)
```

```python
merged_tips_reviews.tail()
```

```python
business_tips_reviews=pd.merge(merged_tips_reviews,business_dataframe,on = 'business_id',how = 'inner')
```

```python
del business_tips_reviews['user_reviews']
```

```python
business_tips_reviews.rename(columns = {'review_count':'total_reviews'})
```

```python
categories_list = business_tips_reviews['categories'].tolist()
broad_categories = []
for i in business_tips_reviews.index:
    category = str(categories_list[i])
    if 'Restaurants' in category or 'Food' in category:
        new_category = 'Eateries'
    elif 'Shopping' in category or  'Hotel' in category or 'Nightlife' in category or 'Entertainment' in category:
        new_category = 'Lifestyle & Entertainment'
    elif 'Medical' in category or 'Active' in category or 'Fitness' in category or 'Beauty' in category:
        new_category = 'Health & Beauty'
    elif 'Auto' in category:
        new_category = 'Auto'
    else:
        new_category = 'Others'
    broad_categories.append(new_category)
business_tips_reviews['broad_categories'] = broad_categories 
```

```python
business_tips_reviews['attributes_Price Range'].fillna(0,inplace=True)
```

```python
business_tips_reviews.rename(columns = {'attributes_Price Range':'Price_Range'})
```
